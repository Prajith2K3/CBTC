```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
unemployment_india_df = pd.read_csv('Unemployment in India.csv')
unemployment_india_df.head(5)
```

```python
unemployment_rate_df = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
unemployment_rate_df.head(5)
```

```python
print(unemployment_india_df.isnull().sum())
```

```python
print(unemployment_rate_df.isnull().sum())
```

```python
unemployment_india_df.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11',
                            'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16',
                            'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19',

'Unnamed: 20', 'Unnamed: 21',
                            'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'], axis=1, inplace=True)
```

```python
unemployment_rate_df.drop(['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13',
                           'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
                           'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23',
                           'Unnamed: 24', 'Unnamed: 25'], axis=1, inplace=True)
```

```python
print(unemployment_india_df['Estimated Unemployment Rate (%)'].describe())
```

```python
unemployment_rate_by_region = unemployment_india_df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
print(unemployment_rate_by_region)
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=unemployment_india_df)

plt.title('Trend of Unemployment Rate in India')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)

plt.show()
```

```python
unemployment_rate_by_region.plot(kind='bar')
plt.title('Unemployment Rate by Region')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Estimated Employed', y='Estimated Unemployment Rate (%)', data=unemployment_india_df)
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Region', y='Estimated Unemployment Rate (%)', data=unemployment_india_df)
plt.title('Box Plot of Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = unemployment_india_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='rainbow', fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
